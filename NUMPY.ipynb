```python
#Horizontal stack
arr1 = np.zeros((2, 3))
arr2 = np.ones((2, 3))

print("Array 1:")
print(arr1)

print("\nArray 2:")
print(arr2)

# Hstack the two arrays
hstacked_array = np.vstack((arr1, arr2))

print("\nHstacked Array:")
hstacked_array
```

```python
# prompt: mathematical functions in numpy

import numpy as np
#Mathematical Functions

arr = np.array([1, 2, 3, 4, 5, 6])
print("Original array:", arr)

# Square root of each element
print("Square root:", np.sqrt(arr))

# Sine of each element
print("Sine:", np.sin(arr))

# Cosine of each element
print("Cosine:", np.cos(arr))

# Exponential of each element
print("Exponential:", np.exp(arr))

# Logarithm (base e) of each element
print("Natural logarithm:", np.log(arr))

# Logarithm (base 10) of each element
print("Logarithm (base 10):", np.log10(arr))

# Absolute value of each element
arr_neg = np.array([-1, -2, 3, -4])
print("Original array (with negative values):", arr_neg)
print("Absolute value:", np.abs(arr_neg))

# Rounding to nearest integer
arr_float = np.array([1.2, 2.7, 3.5, 4.1])
print("Original float array:", arr_float)
print("Rounded:", np.round(arr_float))
print("Floor:", np.floor(arr_float))
print("Ceiling:", np.ceil(arr_float))

# Sum of all elements
print("Sum of all elements:", np.sum(arr))

# Mean of all elements
print("Mean of all elements:", np.mean(arr))

# Standard deviation of all elements
print("Standard deviation:", np.std(arr))

# Variance of all elements
print("Variance:", np.var(arr))

# Minimum element
print("Minimum element:", np.min(arr))

# Maximum element
print("Maximum element:", np.max(arr))

# Index of minimum element
print("Index of minimum element:", np.argmin(arr))

# Index of maximum element
print("Index of maximum element:", np.argmax(arr))

# Cumulative sum
print("Cumulative sum:", np.cumsum(arr))

# Cumulative product
print("Cumulative product:", np.cumprod(arr))

# Differencing
arr_diff = np.array([1, 3, 6, 10])
print("Original array for differencing:", arr_diff)
print("Difference:", np.diff(arr_diff))

# Element-wise comparisons (returns boolean array)
arr_compare1 = np.array([1, 2, 3])
arr_compare2 = np.array([2, 2, 1])
print("Comparison (arr_compare1 > arr_compare2):", arr_compare1 > arr_compare2)

# Logical operations
arr_bool1 = np.array([True, False, True])
arr_bool2 = np.array([False, True, True])
print("Logical AND:", np.logical_and(arr_bool1, arr_bool2))
print("Logical OR:", np.logical_or(arr_bool1, arr_bool2))
print("Logical NOT:", np.logical_not(arr_bool1))

# Unique elements
arr_unique = np.array([1, 2, 2, 3, 3, 3, 4])
print("Original array for unique:", arr_unique)
print("Unique elements:", np.unique(arr_unique))
```

```python
import cv2  #using computer vision
import numpy as np

# Path to your video file
video_path = "/content/VID-20240922-WA0045.mp4"   #Load any video clips in browser folder and give copy path

# Open the video file
cap = cv2.VideoCapture(video_path)

frames = []
while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame = cv2.resize(frame, (128, 128))  # Optional: resize to fixed shape
    frames.append(frame)

cap.release()

# Convert list of frames to a NumPy array
video_array = np.array(frames)

# Shape: (num_frames, height, width, channels)
print("Shape of video array:", video_array.shape)
```

```python
# To get an image (a single frame) from the video array,
# you can use array indexing.
# The video_array has the shape (num_frames, height, width, channels).
# To get a single frame, you need to select an index along the first dimension (num_frames).

# Get the first frame (index 0)
first_frame = video_array[0]
print("\nShape of the first frame:", first_frame.shape)

# Get a specific frame, for example, the 10th frame (index 9)
frame_index = 9
if frame_index < len(video_array):
    tenth_frame = video_array[frame_index]
    print(f"\nShape of the frame at index {frame_index}:", tenth_frame.shape)
else:
    print(f"\nVideo does not have a frame at index {frame_index}.")

# You can then display this frame using libraries like matplotlib or OpenCV
import matplotlib.pyplot as plt

# Display the first frame
plt.imshow(cv2.cvtColor(first_frame, cv2.COLOR_BGR2RGB))
plt.title("First Frame of the Video")
plt.axis('off') # Hide axes
plt.show()

# Display the tenth frame if it exists
if frame_index < len(video_array):
    plt.imshow(cv2.cvtColor(tenth_frame, cv2.COLOR_BGR2RGB))
    plt.title(f"Frame at index {frame_index}")
    plt.axis('off') # Hide axes
    plt.show()
```
